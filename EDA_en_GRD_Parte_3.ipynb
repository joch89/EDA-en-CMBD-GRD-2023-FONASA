## Post 3: Eliminación de filas y creación de columnas


### Eliminación de Valores Nulos

```python
# Eliminar 10 filas con valores nulos en 'FECHA_NACIMIENTO'
df = df.dropna(subset=['FECHA_NACIMIENTO'])

# Eliminar 30 filas con valores nulos en 'IR_29301_COD_GRD'
df = df.dropna(subset=['IR_29301_COD_GRD'])

# Eliminar 172 las filas con valores nulos en 'PROCEDIMIENTO1'
df = df.dropna(subset=['PROCEDIMIENTO1'])
```

### Creación de columnas

- EDAD
- ESTANCIA
- DES, descripción de GRD

#### Columna 'EDAD'

```python
# Convertir 'FECHA_NACIMIENTO' a un objeto tiempo
df['FECHA_NACIMIENTO'] = pd.to_datetime(df['FECHA_NACIMIENTO'], format='%d-%m-%Y', errors='coerce')

# importar fecha y crear una variable que represente la fecha actual.
import datetime
today = datetime.date.today()

# Accede a la data en FECHA_NACIMIENTO y calcula la diferencia en días / 365
df['EDAD'] = (today - df['FECHA_NACIMIENTO'].dt.date).apply(lambda x: x.days) // 365
```

```python
# Mover la columna 'FECHA_NACIMIENTO' a la derecha de la columna 'SEXO'
cols = list(df.columns)
cols.insert(cols.index('SEXO') + 1, cols.pop(cols.index('EDAD')))
df = df[cols]
```

```python
df.head(2)
```

#### Columna ESTANCIA

```python
# Limpieza y arreglo de datos
df['ESTANCIA'] = (pd.to_datetime(df['FECHAALTA'], format='%d-%m-%Y') - pd.to_datetime(df['FECHA_INGRESO'], format='%d-%m-%Y')).dt.days
df['ESTANCIA'] = df['ESTANCIA'].astype(int)
```

#### Columna DESC

Corresponde a la descripción del GRD correspondiente

Creación de la columna entre el cruce de IR_29301_COD_GRD y GRD_DESC

El Archivo excel GRD_DESC contiene una Hoja con dos columnas, una con el código numérico del GRD y otra con la descripción del GRD.

```python
excel_file_path = 'GRD DESC.xlsx'

# cargar archivo excel con columna que contiene la descripción o nombre del GRD correspondiente
grd_desc = pd.read_excel(excel_file_path, sheet_name='Hoja1')
```

```python
# cruzar IR_29301_COD_GRD de df con GRD de grd_desc para crear una columna en df que contenga DESC de grd_desc
df = pd.merge(df, grd_desc, left_on='IR_29301_COD_GRD', right_on='GRD', how='left')
```

```python
# eliminar Columna GRD
df = df.drop('GRD', axis=1)
```

```python
# Mover la columna DESC al lado derecho de IR_29301_PESO
column_to_move = df.pop('DESC')
df.insert(df.columns.get_loc('IR_29301_COD_GRD') + 1, 'DESC', column_to_move)
```

```python
df.head(2)
```

### Cambiar el tipo de dato en columnas

```python
# Convertir las columnas al tipo integer
for col in [ 'IR_29301_SEVERIDAD', 'IR_29301_MORTALIDAD']:
    df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
```

Tipo de datos en DIAGNOSTICO1 y PROCEDIMIENTO1

```python
# Muestra los tipos de datos para las columnas seleccionadas
print(df['DIAGNOSTICO1'].dtype)
print(df['PROCEDIMIENTO1'].dtype)
print(df['IR_29301_COD_GRD'].dtype)
print(df['IR_29301_PESO'].dtype)
print(df['IR_29301_SEVERIDAD'].dtype)
print(df['IR_29301_MORTALIDAD'].dtype)
```

```python
# Remover el .0 de todas las columnas de PROCEDIMIENTO
for i in range(1, 31):
    col_name = f'PROCEDIMIENTO{i}'
    if col_name in df.columns:

        df[col_name] = df[col_name].astype(str)
        df[col_name] = df[col_name].str.replace('.0', '', regex=False)
```

```python
# pasar IR_29301_COD_GRD a numero entero
df['IR_29301_COD_GRD'] = df['IR_29301_COD_GRD'].astype(int)
```

```python
# transformar los NaN de USOPABELLON a 0
df['USOSPABELLON'] = df['USOSPABELLON'].fillna(0)
```

```python
# pasar USOSPABELLON a numero entero
df['USOSPABELLON'] = df['USOSPABELLON'].astype(int)
```

```python
# cambiar la , por . en IR_29301_PESO
df['IR_29301_PESO'] = df['IR_29301_PESO'].str.replace(',', '.')
```

```python
# pasar IR_29301_PESO a float
df['IR_29301_PESO'] = df['IR_29301_PESO'].astype(float)
```

```python
# guardar df_final como csv
df.to_csv('df_final.csv', index=False)
```
